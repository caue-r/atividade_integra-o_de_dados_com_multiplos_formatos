#### Carregando Dados de Clientes (CSV)

```python
dfcsv = pd.read_csv("csv.csv")

dfcsv.head()
```

#### Dados de Pedidos (JSON)

```python
dfjson = pd.read_json("json.json")
dfjson = pd.json_normalize(dfjson["pedidos"])

dfjson.head()
```

### Unindo DataFrames

##### dfjson (dados de pedidos)
##### dfcsv (dados de clientes)
##### dfpipe (dados de avaliação)
##### dfdic (dados de comportamento)

```python
dfjson.head(3)
```

```python
dfcsv.head(3)
```

```python
dfpipe.head(3)
```

```python
dfdic.head(3)
```

```python
df = pd.merge(dfjson, dfcsv, left_on="cliente_id", right_on="id", how="left")
df = pd.merge(df, dfpipe, on=["pedido_id", "cliente_id"], how="left")
df = pd.merge(df, dfdic, on="cliente_id", how="left")
```

```python
df.to_csv("dados_unificados.csv", index=False, encoding="utf-8")
```
